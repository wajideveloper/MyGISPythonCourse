## Tuples

Tuples are immutable sequences, meaning that once a tuple is created, its elements cannot be changed. Tuples are useful for storing fixed collections of items.

For example, a tuple can be used to store the coordinates of a geographic point (latitude, longitude).

```python
tokyo_point = (
    35.6895,
    139.6917,
)  # Tuple representing a geographic point (latitude, longitude)

point_1 = (
    34.6895,
    138.6917
)

print(f"tokyo cordinates:  {tokyo_point}")
print(f"tokyo cordinates new:  {point_1}")
```

```python
# tuple Unpacking
```

You can access elements in a tuple using indexing:

```python
latitude = tokyo_point[0]
longitude = tokyo_point[1]
lat_new = point_1[0]
lon_new = point_1[1]
print(f"Latitude: {latitude}, Longitude: {longitude}")
print(f"New Latitude:\n\t {lat_new}, \t Longitude: {lon_new}")
```

```python
# Different geographic locations as tuples
new_york = (40.7128, -74.0060)
london = (51.5074, -0.1278)
sydney = (-33.8688, 151.2093)
print(f"New York: {new_york}")
print(f"London: {london}")
print(f"Sydney: {sydney}")
```

## Lists

Lists are ordered, mutable sequences, meaning you can change, add, or remove elements after the list has been created. Lists are very flexible and can store multiple types of data, making them useful for various geospatial tasks.

For example, you can store a list of coordinates representing a path or boundary.

Use Lists when:

• You need an ordered collection that can change (like a GPS route or waypoint sequence)

• You want to add, remove, or modify elements dynamically

• Order matters for your application (like steps in a workflow)

9.4.1. Creating Lists

Lists are created using square brackets [] with elements separated by commas. Here are some examples
relevant to geospatial work:


```python
path = [
    (35.6895, 139.6917),
    (34.0522, -118.2437),
    (51.5074, -0.1278),
]  # List of tuples representing a path

print("travel route", path)
```

You can add a new point to the path:

```python
path.append((48.8566, 2.3522))  # Adding Paris to the path
path.append([48.8566, 2.3522])  # Adding Paris to the path
print("Updated path:", path)
```

Lists allow you to perform various operations such as slicing, which lets you access a subset of the list:

```python
sub_path = path[:2]  # Slicing the first two points from the path
print("Sub-path:", sub_path)
```

```python
route = [
(35.6895, 139.6917),  # Tokyo
(34.0522, -118.2437),  # Los Angeles
(51.5074, -0.1278),  # London
]

print(f'that\'s what coordinates are of the route:{route}')
```

## Sets

Sets are unordered collections of unique elements. Sets are useful when you need to store a collection of items but want to eliminate duplicates.

For example, you might want to store a set of unique geographic regions visited during a survey.

Use Sets when:

• You need to eliminate duplicates from your data

• You want to perform operations like finding common elements between datasets

• Order doesn’t matter, but uniqueness does

9.5.1. Creating Sets

You can create sets in several ways. You can use curly braces {} to create a set, or use the set() function
to convert a list or other iterable into a set. Here are examples relevant to geospatial work:

```python
regions = ["North America", "Europe", "Asia"]  # Set of regions
regions = set(regions)
```

```python
country_codes = ["US", "CA", "MX", "US", "CA", "UN"]
unique_codes = set(country_codes)
print("original codes", country_codes)
print("Unique country codes:", unique_codes)
print("length of country code", len(unique_codes))
```

```python
# create a set coordinates sytem
crs_code = {"EPSG:4326", "EPSG:3857", "EPSG:32633"}
print("Coordinate Reference Systems:", crs_code)
```

9.5.2. Adding Elements to Sets

You can add new elements to a set, but duplicates will be automatically ignored:

You can add a new region to the set:

```python
regions.add("Africa")
# regions.append("Shams") set object hasno attributes 'append'
print("Updated regions:", regions)
```

Since sets do not allow duplicates, adding an existing region will not change the set:

```python
regions.add("Europe")  # Attempting to add a duplicate element
print("Regions after attempting to add duplicate:", regions)
```

```python
#adding element tot a set
print("original regions visited..", regions)
regions.add("Antartica")
print("Updated regions visited..", regions)
print("length of the regions...", len(regions))

#try to add a duplicate element to the set
regions.add("Europe")
print("Regions after attempting to add duplicate:", regions)
print("length of the regions...", len(regions))
```

9.5.3. Practical Set Operations

```python
# practice set operation for example

# 1. you might want to find the common countries between two different datasets,
# 2. identify the unique countries in a dataset.
# 3. The intersection() method returns a set of elements that are common to both sets,
# 4. the union()method returns a set of all elements from both sets.


# two different survey areas with their observerd species
area_a_species = {"oak", "pine", "maple", 'birch'}
area_b_species = {"pine", "birch", "cedar", 'fir'}

print("Species in Area A:", area_a_species)
print("Species in Area B:", area_b_species) 
print("\n") 

# Find species common to both areas
common_species = area_a_species.intersection(area_b_species)
print("species found in both areas:", common_species)
print("\n") 

# Find all unique species observed in both areas
unique_to_a = area_a_species.difference(area_b_species) # method 1
unique_to_a_new = area_a_species - area_b_species # method 2
print("Unique: species only in Area A:", unique_to_a)
print("Unique: species only in Area A:", unique_to_a_new)
print("\n\n") 

#find all species across both area
all_species =  area_a_species.union(area_b_species)
print("All species found:", all_species)
```

```python
# Check If we've visited a particular region
if "Asia" in regions:
    print("We have visited Area")

if "Artica" in regions:
    print("We have visited Area")
else:
    print("We have not visited any of these above area")
```

## Dictionaries

Dictionaries are collections of key-value pairs, where each key is unique. Dictionaries are extremely useful for storing data that is associated with specific identifiers, such as attribute data for geographic features.

For example, you can use a dictionary to store attributes of a geospatial feature, such as a city.

9.6.1. Creating Dictionaries

Dictionaries are created using curly braces 
{} with key-value pairs separated by commas and keys
separated from values by colons:

Use Dictionaries when:

• You need to associate descriptive names with values (like feature attributes)

• You want fast lookup of information by key

• You’re organizing complex, structured data about geographic features

```python
tokyo_info = {
    "name": "Tokyo",
    "population": 13929286,
    "coordinates": (35.6895, 139.6917),
    "country" : "Japan",
    "established" : 1603,
}  # Dictionary storing attributes of a city

print("tokyo Information:", tokyo_info)
```

```python
 #Dictionary for a geographic survey point
survey_point = {
"point_id": "SP001",
"latitude": 40.7589,
"longitude": -73.9851,
"elevation": 87.3,
"land_cover": "urban",
"date_surveyed": "2023-06-15",
}
print("Survey point data:", survey_point)
```

9.6.2. Accessing Dictionary Values

You can access the values associated with specific keys:

```python
city_name = tokyo_info["name"]
city_population = tokyo_info["population"]
city_coords = tokyo_info["coordinates"]
print(f" City: {city_name}\n Population: {city_population}\n Coordinates: {city_coords}")
```

9.6.3. Safe Access with get()

The 
get() method provides a safe way to access dictionary values, allowing you to specify a default
value if the key doesn’t exist:

```python
# Safe access to dictionary values
area = tokyo_info.get("area_km2", "Not specified")
timezone = tokyo_info.get("timezone", "Unknown")
print(f"Area: {area}")
print(f"Timezone: {timezone}")
```

```python
area_new = tokyo_info.get("area_meter", "Not specified in this area")
timezone_new = tokyo_info.get("timezone", "Unknown timezone")

print(f"Area:{area_new}")
print(f"TimeZone:{timezone_new}")
```

9.6.4. Adding And Updating Values

You can add New key-value pair or updated exsisting ones:

```python
tokyo_info["area_km2"] = 2194
tokyo_info["timezone"] = "JST"
tokyo_info["population"] = 1400000

print("updated tokyo Info:", tokyo_info)
```

9.6.5. Working with Geographic Feature Collections
Dictionaries are excellent for organizing multiple geographic features:

You can also add or update key-value pairs in a dictionary:

```python
world_capitals = {
    "Japan":{
        "capital":tokyo_info["name"],
        "coordinates":tokyo_info["coordinates"],
        "population":tokyo_info["population"]
    },
    "France": {
"capital": "Paris",
"coordinates": (48.8566, 2.3522),
"population": 2161000,
},
"UK": {
"capital": "London",
"coordinates": (51.5074, -0.1278),
"population": 8982000,
},

}

france_info = world_capitals["France"]
print(f'France\'s capital:{france_info['capital']}')
print(f'Population:{france_info['population']:,}')
```

9.6.6. Dictionary Methods for Data Exploration

Dictionaries provide useful methods for exploring your data, such as 
keys() , 
values() , and 
items() .
These methods return views of the dictionary’s keys, values, and key-value pairs, respectively. They are
useful for iterating over the dictionary and for checking if a key exists in the dictionary. We will learn
more about these methods in the Loops and Conditional Statements chapter. Below is an example of
how to use these methods.

```python
print("Keys in Tokyo_info:", list(tokyo_info.keys()))
print("Keys in Tokyo_info:", list(tokyo_info.values()))
```

```python
if "coordinates" in tokyo_info:
    print("coordinates infromation is avalible")

countries = list(world_capitals.keys())
print("Countries in our database:", countries)
```
